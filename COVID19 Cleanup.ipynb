```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid19.csv')
```

```python
df['Province | State'] = df['Province/State'].fillna(df['Province_State'])
```

```python
df['Country | Region'] = df['Country/Region'].fillna(df['Country_Region'])
```

```python
df['Last Update'] = df['Last Update'].fillna(df['Last_Update'])
```

```python
df['Lat'] = df['Lat'].fillna(df['Latitude'])
df['Long'] = df['Long_'].fillna(df['Longitude'])
```

```python
df.drop(columns=['Province/State', 'Country/Region', 'Last_Update', 'Province_State', 'Country_Region',
                 'Longitude', 'Long_', 'Latitude'], inplace=True)
```

```python
df = df[['Country | Region', 'Province | State', 'Active', 'Confirmed', 'Deaths', 'Recovered', 'FIPS', 'Admin2', 'Lat',
   'Long', 'Combined_Key', 'Last Update']]
```
